### Задание 1.

Дан объект в $2D$ пространстве

```python
import numpy as np
import matplotlib.pyplot as plt

A = [[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]]

A = np.array(A)
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

При помощи линейных отображений:
* Уменьшить объект в два раза
* Повернуть на 130 градусов
* Отразить объект относительно прямой y=x

Затем вычислить определитель композиции преобразований (итоговой матрицы реализующей все три отображения)

```python
# Поскольку фактически из исходной матрицы 6x3 нам надо производить манипуляции с матрицей 6x2, вначале выделим ее
B = A[:,0:2]

# Далее транспонируем матрицу, чтобы из матрицы 6x2 получить матрицу 2x6
C0 = B.transpose()

# Теперь исходя из условий задачи формируем матрицы перехода
# Первая матрица уменьшит объект в 2 раза:
M1 = np.array([[0.5, 0],[0, 0.5]])

# Вторая матрица повернет объект на 130 градусов
# Для угла 130 градусов считаем косинус и синус, предварительно переведя градусы в радианы:
cos130 = np.cos(np.deg2rad(130))
sin130 = np.sin(np.deg2rad(130))
M2 = np.array([[cos130, -sin130],[sin130, cos130]])

# Третья матрица отразит объект относительно прямой y=x (используем обратнодиагональную единичную матрицу)
M3 = np.array([[0, 1],[1, 0]])

# Таким образом "на старте" имеем следующие матрицы:
print('Исходная матрица:\n', C0)
print('\nПервая матрица перехода:\n', M1)
print('\nВторая матрица перехода:\n', M2)
print('\nТретья матрица перехода:\n', M3)
```

```python
# Последовательно применяем матрицы перехода и рисуем графики исходной и преобразованных фигур
C1 = M1@C0
C2 = M2@C1
C3 = M3@C2

# Для наглядности меняем стандартный размер диаграммы, делая ее квадратной
plt.figure(figsize=(8,8))
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.plot(C0[0], C0[1])
plt.plot(C1[0], C1[1])
plt.plot(C2[0], C2[1])
plt.plot(C3[0], C3[1])

plt.legend(['Исходная фигура', 'Масштабирование', 'Поворот 130', 'Отражение'])
```

```python
# Дополнительно можно попробовать соорудить из трех матриц преобразования одну, 
# которая одновременно выполнит все три операции.
 
M4 = M3@(M2@M1)
print('Универсальная матрица преобразования:\n',M4)

# Применяем к исходной матрице универсальную матрицу преобразования
C4 = M4@C0

# Смотрим, что получилось: 
plt.figure(figsize=(8,8))
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.plot(C0[0], C0[1])
plt.plot(C4[0], C4[1])
plt.legend(['Исходная фигура', 'Результат'])
```

### Задание 2.

Решите систему уравнений, используя матричный подход. Необходимо составить матрицу коэффициентов и вектор правых частей. 

$x + z  = 2$

$-x + y - 2*z + w = -2$

$4*x + z - 2*w = 0$

$-4*x + 4*y + w = 5$


```python
# Фактически мы имеем дело с системой из четырех уравнений с четырьмя неизвестными.
# Перепишем уравнения следующим образом:
# 1*x + 0*y + 1*z + 0*w = 2
# -1*x + 1*y - 2*z + 1*w = -2
# 4*x + 0*y + 1*z - 2*w = 0
# -4*x + 4*y + 0*z + 1*w = 5

# Теперь легко сформировать исходные матрицы.
# Матрица коэфициентов
M1 = np.array([[1,0,1,0],[-1,1,-2,1],[4,0,1,2],[-4,4,0,1]])

# Также создадим вектор со значениями из правых частей уравнений
B = np.array([[2],[-2],[0],[5]])

# Для получения решения вначале нам надо найти обратную матрицу для матрицы коэффициентов
M2 = np.linalg.inv(M1)

# Выведем все три матрицы для визуального контроля :)
print('Матрица коэффициентов:\n',M1)
print('Обратная матрица коэффициентов:\n',M2)
print('"Правый" вектор:\n', B)
```

```python
# Проверяем корректность полученной обратной матрицы
print(M1@M2)

# Пытаемся решить уравнение, умножив обратную матрицу коэффициентов на "правую" матрицу
R = M2@B
print('Искомые x, y, z, w:\n', R)

# Пробуем прийти к тому же решению с использованием функции solve из библиотеки numpy linalg
R2 = np.linalg.solve(M1,B)
print('Искомые x, y, z, w:\n', R2)
```

### Задание 3.

Найдите спектральное разложение матрицы

```python
m = np.array([[1, 2],
              [2,3]])
```

```python
# Спектральное разложение возможно для симметричных матриц. В нашем случае двумерная матрица m является симметричной.
# Мы должны в результате получить три красивых матрицы, произведение которых позволяет получить исходную матрицу.
# Одна из этих матриц - диагональная, вторая состоит из набора ортогональных векторов, а третья - из транспонированной матрицы
# тех же ортогональных векторов.
# m = St * D * S, где D - диагональная, а S - ортогональная

# Для решения этого задания нам фактически нужно для начала найти собственные числа и собственные вектора исходной матрицы
# Собственные числа позволят получить диагональную матрицу, а собственные вектора позволят получить ортогональную матрицу.
self_numbers, self_vectors = np.linalg.eigh(m)
print(f'Собственные числа:\n{self_numbers},\nсобственные вектора:\n{self_vectors}')
```

```python
# Используем функцию библиотеки numpy для создания диагональной квадратной матрицы из вектора, содержащего значения диагонали
D = np.diag(self_numbers)

# Ортогональную матрицу построим из комплекта двух собственных векторов
S = self_vectors.transpose()
St = S.transpose()

# Посмотрим на получившиеся матрицы:
print(f'Матрица D:\n{D},\nМатрица S:\n{S},\nМатрица St:\n{St}')
```

```python
# В нашем случае получилось так, что матрицы S и St - тоже симметричные, поэтому они равны друг другу.
# Проверим, что получается в результате умножения получившихся матриц
m2 = St@(D@S)
print(f'Результат умножения матриц, полученных в результате спектрального разложения:\n{m2}')
```
