```python
# Импортируем библиотеки
import numpy as np
import matplotlib.pyplot as plt
# Для манипуляций с функциями и их производными
import sympy as sym
```

### Задание 1.

Вычислить значения первой и второй производной функции:

$f(x)=x^5+4sin(2x)+cos(3x+3)$

в точке $x = 1$.

```python
# определяем x как символ (переменную функции)
x = sym.Symbol('x')
# задаем формулу, используя определения функций sin и cos из библиотеки sym
src = x**5 + 4*sym.sin(2*x) + sym.cos(3*x + 3)
src
```

```python
# вычисляем первую и вторую производные функции
res_1 = sym.diff(src,x)
print('Первая производная: ', res_1)
res_2 = sym.diff(res_1,x)
print('Вторая производная: ', res_2)
```

```python
# Выводим результаты. Поскольку subs производит просто замену переменной X значением 1, 
# для вычисления значения функции после замены переменной вызываем метод evalf()
print(f'В точке х=1 значение первой производной = {res_1.subs(x,1).evalf()}')
print(f'значение второй производной = {res_2.subs(x,1).evalf()}')
```

#### Функция ReLU

Популярная и простая функция активации - [функция ReLU](https://ru.wikipedia.org/wiki/%D0%A4%D1%83%D0%BD%D0%BA%D1%86%D0%B8%D1%8F_%D0%B0%D0%BA%D1%82%D0%B8%D0%B2%D0%B0%D1%86%D0%B8%D0%B8). Преимущество данной функции - ее математический вид. Функция требует немного вычислительных ресурсов, что ускоряет физическое время обучения нейронной сети, это хорошо подходит для создания глубинных нейронных сетей.

```python
def relu(x):
    return np.maximum(0.0, x)
```

```python
x = np.arange(-3, 3, 0.1)
y = relu(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('ReLU Function')
plt.show()
```

```python
def relu_derivative(x):
    # Вычислить производную функции ReLU
    # Ваш код напишите ниже и результат передайте в переменную derivative
    # Почему-то хочется "тупо" решить эту задачу, просто проанализировав график функции
    # Очевидно, что от -бесконечности до 0 производная равна нулю (касательная к прямой совпадает с прямой,
    # тангенс угла наклона при угле 0 градусов равен также 0)
    # Столь же очевидно, что 0 - точка излома, и производная в ней не определена.
    # Если же x>0, то фактически y=x, это - прямая, идущая под углом 45 градусов, тангенс равен 1 (как и производная)    
    if x <= 0:
        derivative = 0
    else:
        derivative = 1
    
    return derivative
```

```python
# Для более точного построения уменьшаю шаг дискретизаци на порядок
x = np.arange(-3, 3, 0.01)
# Получаем набор значений производной y_der для заданного диапазона значений x
der_func = np.vectorize(lambda x_val: relu_derivative(x_val))
y_der = der_func(x)

# Строим графики исходной функции и её производной
plt.plot(x, y)
plt.plot(x, y_der)
plt.xlabel('x')
plt.ylabel('y')
plt.title('ReLU Function and it derivative')
plt.show()
```

#### Сигмоидная функция

Популярная функция активации - [сигмоидная логистическая функция](https://ru.wikipedia.org/wiki/%D0%A1%D0%B8%D0%B3%D0%BC%D0%BE%D0%B8%D0%B4%D0%B0). Сигмоида очень часто применяется в задачах классификации, так как является  гладкой функцией, а область значений ограничена от $-1$ до $1$. Еще одним достоинством логистической функции является вид ее производной, который позволяет существенно сэкономить вычислительные ресурсы. 

Напишите код для определения производной сигмоиды и постарайтесь понять почему данная функция активации практически не требует дополнительных вычислительных ресурсов при расчете значений производной. 

```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-5, 5, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sigmoid Function')
plt.show()
```

```python
def sigmoid_derivative(x):
    # Вычислить производную сигмоидной функции
    # Ваш код напишите ниже и результат передайте в переменную derivative
    # Здесь можно воспользоваться линейной аппроксимацией функции на минимальном диапазоне 
    # и вычислением тангенса угла наклона этой линии
    # Берем маленький отрезок, заведомо маленький по сравнению со скоростью изменения функции
    delta = 0.0001
    # Определяем координаты X двух точек 
    xx = np.array([x-delta, x+delta])
    # Вычисляем соответствующие им координаты Y
    yy = sigmoid(xx)
    # вычисляем dy/dx, фактически и определяющее производную
    derivative = (yy[1] - yy[0])/(delta*2);    
    return derivative
```

```python
x = np.arange(-5, 5, 0.1)
#y = sigmoid(x)
# Получаем набор значений производной y_der для заданного диапазона значений x
der_func2 = np.vectorize(lambda x_val: sigmoid_derivative(x_val))
y_der2 = der_func2(x)

plt.plot(x, y, label="Исходная функция")
plt.plot(x, y_der2, label="Производная")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Sigmoid Function and its Derivative')
plt.show()
```
